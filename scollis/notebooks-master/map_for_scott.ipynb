```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
min_lat=-12.7
max_lat=-12.2
min_lon=130.5
max_lon=131.3
ax=fig.add_axes([0.1,0.1,0.8,0.8])
m = Basemap(llcrnrlon=min_lon,llcrnrlat=min_lat,
                    urcrnrlon=max_lon,urcrnrlat=max_lat,
                    projection='mill',area_thresh=10,
                    resolution='h')

runway1x,runway1y=m(array([130.865321, 130.89427]),array([-12.40945, -12.419184]))
runway2x,runway2y=m(array([130.870643, 130.870696]), array([-12.409041, -12.422505]))
apannotx, apannoty=m(130.75, -12.37)
armx, army=m(130.891848,-12.424575)
cpolx, cpoly=m(131.044402,-12.249027)
bayx, bayy=m(130.9,-12.3)
m.drawcoastlines()
m.fillcontinents()
m.drawmapboundary(fill_color='blue')
m.drawparallels(linspace(-11, -13, 11),labels=[1,0,0,0])
m.drawmeridians(linspace(130,132, 11),labels=[0,0,0,1])
m.drawrivers()
m.plot(runway1x, runway1y, 'k-')
m.plot(runway2x, runway2y, 'k-')
# P.arrow( x, y, dx, dy, **kwargs )
#P.arrow( 0.5, 0.8, 0.0, -0.2, fc="k", ec="k",
arrow(apannotx, apannoty, runway1x[0]-apannotx-4000, 
      runway1y[0]-apannoty+1000, head_width=3000, 
      head_length=3000, fc='w', ec='w' )
text(apannotx-5000,apannoty,'Airport', fontdict={'color':'white', 'weight':'bold'})
m.plot(armx,army,'bo')
text(armx+1500, army-3000, 'ARM Darwin\n site', fontdict={'color':'black', 'weight':'bold'})
m.plot(cpolx,cpoly,'bo')
text(cpolx+1500, cpoly+1000, 'CPOL Radar', fontdict={'color':'black', 'weight':'bold'})
text(bayx,bayy,'Shoal \n Bay', fontdict={'color':'white', 'weight':'bold'})
m.drawmapscale(130.7, -12.3, 130.6, -12.3, 20, barstyle='fancy', fontcolor='w', fillcolor1='w', fillcolor2='k')

ax2=fig.add_axes([0.63,0.2,0.2,0.3])
#[*left*, *bottom*, *width*,*height*]
min_lat=-17
max_lat=-5
min_lon=125.
max_lon=137.
m2 = Basemap(llcrnrlon=min_lon,llcrnrlat=min_lat,
                    urcrnrlon=max_lon,urcrnrlat=max_lat,
                    projection='mill',area_thresh=10,
                    resolution='h')

#m2.drawcoastlines()
#m2.fillcontinents()
#m2.drawmapboundary(fill_color='blue')
m2.etopo()
gca().axes.get_xaxis().set_visible(False)
gca().axes.get_yaxis().set_visible(False)
armx, army=m2(130.891848,-12.424575)

m2.plot(armx,army,'bo')

savefig('/Users/scollis/Desktop/TWP_map.png', dpi=300)
```
