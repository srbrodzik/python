```python
def mslp_and_relf(data, radar_data, **kwargs):
    #map ranges
    min_lon=kwargs.get('min_lon', -92)
    max_lon=kwargs.get('max_lon', -86)
    min_lat=kwargs.get('min_lat', 40)
    max_lat=kwargs.get('max_lon', 44)
    time_step=kwargs.get('time_step', 0)
    z_lev=kwargs.get('z_lev', 0)
    axes_loc=kwargs.get('axes_loc',[0.05,0.05,0.9,0.80])
    lat_lines=kwargs.get('lat_lines', np.arange(38,46,2))
    lon_lines=kwargs.get('lon_lines', np.arange(-94,-80,2))
    # contour levels
    clevs = kwargs.get('clevs', np.arange(900,1100.,1.))
    # read lats,lons.
    lats = data.variables['lat'][:]
    lons = (360.0-data.variables['lon'][:])*-1.0
    nlats = len(lats)
    nlons = len(lons)
    # read prmsl, convert to hPa (mb).
    prmsl = 0.01*data.variables[kwargs.get('msl','prmsl')][time_step]
    # the window parameter controls the number of highs and lows detected.
    # (higher value, fewer highs and lows)
    print prmsl.min(), prmsl.max()
    local_min, local_max = extrema(prmsl, mode='wrap', window=50)
    # create Basemap instance.
    m =\
    Basemap(llcrnrlon=min_lon,llcrnrlat=min_lat,
            urcrnrlon=max_lon,urcrnrlat=max_lat,
            projection='mill',area_thresh=10000,
            resolution='l')
    # find x,y of map projection grid.
    lons, lats = np.meshgrid(lons, lats)
    x, y = m(lons, lats)
    # create figure.
    fig=plt.figure(figsize=(10,5))
    ax = fig.add_axes(axes_loc) 
    #loop over radars and plot
    xd,yd=m(radar_data[0], radar_data[1])
    pcmesh=m.pcolormesh(xd,yd,radar_data[2][z_lev,:,:] , vmin=-8, vmax=64)
    cs = m.contour(x,y,prmsl,clevs,colors='k',linewidths=1.)
    #contour labeling
    class nf(float):
         def __repr__(self):
             str = '%.1f' % (self.__float__(),)
             if str[-1]=='0':
                 return '%.0f' % self.__float__()
             else:
                 return '%.1f' % self.__float__()
    
    # Recast levels to new class
    cs.levels = [nf(val) for val in cs.levels ]
    
    # Label levels with specially formatted floats
    if plt.rcParams["text.usetex"]:
         fmt = r'%r hPa'
    else:
         fmt = '%r hPa'
    plt.clabel(cs, cs.levels, inline=True, fmt=fmt, fontsize=10)
    
    m.drawcoastlines(linewidth=1.25)
    m.drawstates()
    m.drawparallels(lat_lines,labels=[1,1,0,0])
    m.drawmeridians(lon_lines,labels=[0,0,0,1])
    xlows = x[local_min]; xhighs = x[local_max]
    ylows = y[local_min]; yhighs = y[local_max]
    lowvals = prmsl[local_min]; highvals = prmsl[local_max]
    # plot lows as blue L's, with min pressure value underneath.
    xyplotted = []
    # don't plot if there is already a L or H within dmin meters.
    yoffset = 0.022*(m.ymax-m.ymin)
    dmin = yoffset
    for x,y,p in zip(xlows, ylows, lowvals):
        if x < m.xmax and x > m.xmin and y < m.ymax and y > m.ymin:
            dist = [np.sqrt((x-x0)**2+(y-y0)**2) for x0,y0 in xyplotted]
            if not dist or min(dist) > dmin:
                plt.text(x,y,'L',fontsize=14,fontweight='bold',
                        ha='center',va='center',color='b')
                plt.text(x,y-yoffset,repr(int(p)),fontsize=9,
                        ha='center',va='top',color='b',
                        bbox = dict(boxstyle="square",ec='None',fc=(1,1,1,0.5)))
                xyplotted.append((x,y))
    # plot highs as red H's, with max pressure value underneath.
    xyplotted = []
    for x,y,p in zip(xhighs, yhighs, highvals):
        if x < m.xmax and x > m.xmin and y < m.ymax and y > m.ymin:
            dist = [np.sqrt((x-x0)**2+(y-y0)**2) for x0,y0 in xyplotted]
            if not dist or min(dist) > dmin:
                plt.text(x,y,'H',fontsize=14,fontweight='bold',
                        ha='center',va='center',color='r')
                plt.text(x,y-yoffset,repr(int(p)),fontsize=9,
                        ha='center',va='top',color='r',
                        bbox = dict(boxstyle="square",ec='None',fc=(1,1,1,0.5)))
                xyplotted.append((x,y))
    #plt.title('Mean Sea-Level Pressure (with Highs and Lows) %s' % date)
    cbax = fig.add_axes([0.2,0.95,0.6,0.03])
    
    cbar=plt.colorbar(mappable=pcmesh, cax=cbax, orientation='horizontal')
    cbar.set_label(r'Eq. Refl. Fac. (dBZ)')
    plt.savefig(kwargs.get('fname', '/Users/scollis/testmemsl.png'), dpi=300)

"""
plot H's and L's on a sea-level pressure map
(uses scipy.ndimage.filters and netcdf4-python)
"""
def extrema(mat,mode='wrap',window=10):
    """find the indices of local extrema (min and max)
    in the input array."""
    mn = minimum_filter(mat, size=window, mode=mode)
    mx = maximum_filter(mat, size=window, mode=mode)
    # (mat == mx) true if pixel is equal to the local max
    # (mat == mn) true if pixel is equal to the local in
    # Return the indices of the maxima, minima
    return np.nonzero(mat == mn), np.nonzero(mat == mx)

def my_qrf(xg, yg, zg):
    return 1200.0

    cf_lat = dms_to_d([36.0, 36.0, 18.35])
    cf_lon = -1.0 * dms_to_d([97.0, 29.0, 10.69])

    # perform gridding
    grids = pyart.map.map_to_grid(
        (radar,),
        grid_shape=(241, 241, 2),
        grid_limits=((-117000.0, 123000.0), (-100000.0, 140000.0),
                     (1500, 2000)),
        grid_origin=(cf_lat, cf_lon),
        fields=['reflectivity_horizontal', 'copol_coeff'],
        toa=20000.0,
        qrf_func=my_qrf,
        copy_field_data=True)

def map_prep_radar(filenames):
    radars=[pyart.io.read_netcdf(filen) for filen in filenames]
    #radars[0].time.update({'calendar':'gregorian'})
    grids = pyart.map.grid_from_radars(
        radars,
        grid_shape=(341, 341, 2),
        grid_limits= ((-350.00*1000.0, +250.*1000.0), (-400.0*1000.0, 300.0*1000.0),(500, 1000)),
        fields=['REF'],
        refl_field='REF',
        max_refl=100., grid_origin=(41.0, -89.0), qrf_func=my_qrf,copy_field_data=True)
    lat = 41.0
    lon = -89
    pnyc = pyproj.Proj(proj='lcc',datum='NAD83',lat_0=lat,lon_0=lon,x_0=0.0,y_0=0.0)
    lon, lat = pnyc(grids.axes['x_disp']['data'], grids.axes['y_disp']['data'], inverse=True)
    Ze=grids.fields['REF']['data']
    return lon, lat, Ze

def fetch_and_convert_nexrad(url, tmp_dir='/data/tmp/'):
    last_file=urllib.urlopen(url+'dir.list').read().split('\n')[-3].split(' ')[-1]
    outfile=tmp_dir+last_file
    req = urllib2.urlopen(url+last_file)
    outobj=open(outfile, 'wb')
    outobj.write(req.read())
    outobj.close()
    command="/usr/local/bin/RadxConvert -const_ngates -outdir "+ tmp_dir+" -f " + outfile
    os.system(command)
    dump_dir=tmp_dir+last_file.split('_')[1]
    files=os.listdir(dump_dir)
    ofilename=last_file.split('_')[0]+'_latest.nc'
    os.system('mv '+dump_dir+'/'+files[0]+' /data/tmp/'+ofilename)
    os.rmdir(dump_dir)
    os.remove(outfile)
    return ofilename
```
