```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.basemap import Basemap

import json
import pyart
import matplotlib 
%matplotlib inline

font = {'size': 12}
matplotlib.rc('font', **font)
```

```python
my_file = '/data/KVNX20110520_115851_V06'
my_radar = pyart.io.read(my_file)
```

```python
def map_json(radar, json_filename, figsize = None, var ='reflectivity',
             vmin = None, vmax = None, resolution = 'l', sweep = 0):
    """map_json
    docutils to follow
    """
    json_file_handle = open(json_filename)
    js_dict = json.loads(json_file_handle.read())
    json_file_handle.close()
    
    if 'lat_spacing' in js_dict.keys():
        lat_lines = np.arange(js_dict['min_lat'],js_dict['max_lat'], js_dict['lat_spacing'])
    else:
        lat_lines = None
    if 'lon_spacing' in js_dict.keys():
        lon_lines = np.arange(js_dict['min_lon'],js_dict['max_lon'], js_dict['lon_spacing'])
    else:
        lon_lines = None
    
    f = plt.figure(figsize = figsize)
    pyart_map_display = pyart.graph.RadarMapDisplay(radar)
    pyart_map_display.plot_ppi_map( 'reflectivity', sweep = sweep, max_lat = js_dict['max_lat'],
                         min_lat = js_dict['min_lat'], min_lon = js_dict['min_lon'], 
                         max_lon = js_dict['max_lon'], vmin = vmin,
                         vmax = vmax, lat_lines = lat_lines,
                         lon_lines = lon_lines,
                         resolution = resolution)
    
    if 'layers' in js_dict.keys():
        layer_dict = js_dict['layers']
        for layer in ['states', 'rivers']:
            if layer in layer_dict.keys():
                if layer == 'states':
                    pyart_map_display.basemap.drawstates(**layer_dict[layer])
                if layer == 'rivers':
                    pyart_map_display.basemap.drawrivers(**layer_dict[layer])
    if 'labels' in js_dict.keys():
        label_dict = js_dict['labels']
        for i in range(len(label_dict['text'])):
            pyart_map_display.plot_point(label_dict['locations'][i][1], label_dict['locations'][i][0],
                       symbol = 'ro', label_text = label_dict['text'][i], 
                       label_offset=(label_dict['offsets'][i][0], label_dict['offsets'][i][1]))
    return pyart_map_display
```
