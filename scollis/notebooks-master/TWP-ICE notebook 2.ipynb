First we read in the cube to be analyzed and a fairly rainfall free field to nuke the clutter.. in an ideal case we would nuke the clutter before we map to a grid, but here we do it post mapping.

```python
mygrids=pyart.io.grid.read_grid('/data/grid_cfrad.20060119_005009.000_to_20060119_005810.000_CPOL_v1_PPI..nc')
myclutter=pyart.io.grid.read_grid('/data/grid_cfrad.20060119_075009.000_to_20060119_075810.000_CPOL_v1_PPI..nc')
```

The clutter volume has some legit returns so we restrict our study to an area of known clutter around the Tiwi's and set the rain rate there to zero. 

```python
clt=np.zeros(myclutter.fields['rain_rate_A']['data'].shape)
region=np.zeros(clt.shape)
region[:,120:190, 0:75]=1.0
iscluttered=np.where(myclutter.fields['rain_rate_A']['data']*region > 0.1)
mygrids.fields['rain_rate_A']['data'][iscluttered]=0.0
```

Now we run the reflectivity field through the Steiner code (Steiner and Houze 1997) which returns a 2D grid where 1.0 means tagged convective and -1 means non-convective (so rain free or stratiform). NOTE: This is tuned for Oklahoma, but again, this is illustrative only. We would need to tune the algorithm using distrometer data (available) to ensure we got good microphysical seperation using the texture.

Steiner, M., and R. A. Houze, 1997: Sensitivity of the Estimated Monthly Convective Rain Fraction to the Choice of   Z   –   R   Relation. J. Appl. Meteor., 36, 452–462, doi:10.1175/1520-0450(1997)036<0452:SOTEMC>2.0.CO;2.


```python
eclass = echo_class.steiner(mygrids.axes['x_disp']['data'], 
                 mygrids.axes['y_disp']['data'], 
                 mygrids.axes['z_disp']['data'], 
                 mygrids.fields['corrected_reflectivity_horizontal']['data'],
                 dx=500.0, dy=500.0, intense=45.0, bkg_rad=11000.0, 
                 area_relation='sgp', work_lev=1500.0, peak_relation='sgp', 
                 use_intense=True, fill_val=-9999.0)
```

Now lets make a pretty display

```python
display = pyart.graph.GridMapDisplay(mygrids, debug=True)

fig = plt.figure(figsize=[15, 8])

# panel sizes
map_panel_axes = [0.05, 0.05, .4, .80]
x_cut_panel_axes = [0.55, 0.10, .4, .30]
y_cut_panel_axes = [0.55, 0.50, .4, .30]
colorbar_panel_axes = [0.05, 0.90, .4, .03]

# parameters
level = 0
vmin = 0.10
vmax = 180
lat = -11.9
lon = 130.7

# panel 1, basemap, radar reflectivity and NARR overlay
ax1 = fig.add_axes(map_panel_axes)
display.plot_basemap(lat_lines=np.arange(-11.5, -13.5, -.2) , 
                     lon_lines=np.arange(129,132,.2), area_thresh=100,
                     resolution='h')
display.plot_grid('rain_rate_A', level=level, vmin=vmin, vmax=vmax)
display.plot_crosshairs(lon=lon, lat=lat)
xd, yd = display.basemap(display.grid_lons, display.grid_lats)
display.basemap.contour(xd, yd, eclass, [0.5], colors=['r'], linewidths=1.)

# colorbar
cbax = fig.add_axes(colorbar_panel_axes)
display.plot_colorbar()

# panel 2, longitude slice.
ax2 = fig.add_axes(x_cut_panel_axes)
display.plot_longitude_slice('rain_rate_A', lon=lon, lat=lat, vmin=vmin, vmax=vmax)
ax2.set_xlabel('Distance from SGP CF (km)')

# panel 3, latitude slice
ax3 = fig.add_axes(y_cut_panel_axes)
display.plot_latitude_slice('rain_rate_A', lon=lon, lat=lat, vmin=vmin, vmax=vmax)

# add a title
slc_height = mygrids.axes['z_disp']['data'][level]
dts = num2date(mygrids.axes['time']['data'], mygrids.axes['time']['units'])
datestr = dts[0].strftime('%H:%M Z on %Y-%m-%d')
title = 'Sliced at ' + str(slc_height) + ' meters at ' + datestr
fig.text(0.5, 0.9, title)
```

Now to extract some info.. The Numerical Python "Where" command <a href="http://docs.scipy.org/doc/numpy/reference/generated/numpy.where.html"> (Click here)</a> is a very useful command for conditional sampling.. Here we use it to create conditional means on convective, non-convective elements.. as well as getting a count of pixels 

```python
mean_rr = mygrids.fields['rain_rate_A']\
                        ['data'][1,:,:].mean()
conv_mean = mygrids.fields['rain_rate_A']\
                        ['data'][1,np.where(eclass == 1.0)].mean()
strat_mean = mygrids.fields['rain_rate_A']\
                        ['data'][1,np.where(eclass == -1.0)].mean()
npts_rain = len(np.where(mygrids.fields['rain_rate_A']\
                        ['data'][1,:,:] >0.1)[0])
npts_conv = len(np.where(eclass == 1.0)[0])
```

```python
time=num2date(mygrids.axes['time']['data'], mygrids.axes['time']['units'])
print(time)
print("Mean rain rate: %f mm/h\nStratiform mean: %f mm/h\nConvective mean: %f mm/h\nArea fraction raining: %f\nArea fraction convecitive: %f"\
      % (mean_rr, strat_mean, conv_mean, 
         float(npts_rain)/(eclass.shape[0]*eclass.shape[1]), 
         float(npts_conv)/(eclass.shape[0]*eclass.shape[1]))) 
```
