```python
import matplotlib.pyplot as plt
import sys
import os
pyart_dir=os.environ.get('PYART_DIR',os.environ['HOME']+'/python')
sys.path.append(pyart_dir)
from pyart.io import  radar
from pyart.graph import rayplot, radar_display
from pyart.io import nc_utils
import netCDF4
from copy import deepcopy
import numpy as np
from matplotlib import animation

def dt_to_dict(dt, **kwargs):
    pref = kwargs.get('pref', '')
    return dict([(pref+key, getattr(dt, key)) for key in
                ['year', 'month', 'day', 'hour', 'minute', 'second']])

```

```python
files=os.listdir('/data/')
want_files=[]
for file in files:
    if 'sgpxsaprsesurcmacI4.c0.20110520' in file:
        want_files.append('/data/'+file)
my_nc_object=netCDF4.Dataset(want_files[0], 'r')
myradar=radar.Radar(my_nc_object)
my_nc_object.close()
```

```python
want=['reflectivity_horizontal', 'copol_coeff']
myrayplot=rayplot.rayplot(myradar, figsize=[15,15])
myrayplot([0,300],want)
myrayplot.fig.subplots_adjust(wspace=1.0)
```
