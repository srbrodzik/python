```python
#first we do some imports and check the version of Py-ART for consistency
import pyart
from matplotlib import pyplot as plt
import netCDF4
import numpy as np
from copy import deepcopy
import os
import scipy
%matplotlib inline
print pyart.__version__
```

```python
mydir = '/data/vdis/' 
all_files = os.listdir(mydir)
good_files = []
for fl in all_files:
    if ".cdf" in fl:
        good_files.append(mydir +fl)
good_files.sort()
```

```python
ncobj = netCDF4.MFDataset(good_files)
```

```python
print ncobj.variables.keys()
print ncobj.variables['time'].units
print ncobj.variables['rain_rate'].units
```

```python
time_object_array = netCDF4.num2date(ncobj.variables['time'][:], ncobj.variables['time'].units)
```

```python
f = plt.figure(figsize = [15,5])
plt.plot_date(time_object_array, ncobj.variables['rain_rate'][:], 'r-')
```

```python
f = plt.figure(figsize = [15,5])
plt.plot(ncobj.variables['time'][:])
```

Ok, you can see the problem here is ARM starndards have time as time since midnight, and NetCDF4's default is that metadata for all variables is the same.. So we are going to do a crude trick here using np.unfold


```python
good_time = np.unwrap(ncobj.variables['time'][:], discont = 80000)
good_time_object_array = netCDF4.num2date(good_time, ncobj.variables['time'].units)
```

```python
f = plt.figure(figsize = [15,5])
plt.plot(good_time)
```

```python
f = plt.figure(figsize = [15,5])
plt.plot_date(good_time_object_array, ncobj.variables['rain_rate'][:], 'r-')
plt.xlabel('Date')
plt.ylabel('Rain Rate')
```

```python
dt = good_time[1] - good_time[0]
accum = np.cumsum(ncobj.variables['rain_rate'][:]) * (dt / (60.*60.))
```

```python
print accum.shape, good_time_object_array.shape
```

```python
f = plt.figure(figsize = [15,5])
plt.plot_date(good_time_object_array, ncobj.variables['rain_rate'][:], 
              'r-', label = 'Rain Rate')
plt.xlabel('Date')
plt.ylabel('Rain Rate (mm/hr)')
plt.twinx()
plt.plot_date(good_time_object_array, accum, 
              'b-', label = 'Accum. Rainfall')
plt.ylabel('Rain accumilation (mm)')

```
