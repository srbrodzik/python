```python
#imports
import pyart
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from time import time
```

```python
myradar = pyart.io.read('/data/KLOT_20150410_0056')
```

determine which sweeps have no data

```python
good = []
for i in range(myradar.nsweeps):
    sweep = myradar.get_slice(i)
    if myradar.fields['velocity']['data'][sweep][0, :].flatten().count() != 0:
        good.append(i)
```

```python
myradar = myradar.extract_sweeps(good)
```

Gates in velocity that are masked

```python
gatefilter = pyart.correct.GateFilter(myradar)
gatefilter.exclude_masked('velocity')
```

```python
corr_vel = pyart.correct.dealias_region_based(
    myradar, vel_field='velocity', keep_original=False, gatefilter = gatefilter, nyquist_vel=26.)
myradar.add_field('corrected_velocity', corr_vel, True)
```

```python
my_display = pyart.graph.RadarDisplay(myradar)
tilt = 0
f = plt.figure(figsize=[20,6])
plt.subplot(1,2,1)
my_display.plot_ppi('reflectivity', tilt, vmin=-6, vmax=64.0)
ax = plt.gca()
ax.set_xlim([-60,0])
ax.set_ylim([30,100])
plt.subplot(1,2,2)
my_display.plot_ppi('velocity', tilt, vmin=-26., vmax=26.)
ax = plt.gca()
ax.set_xlim([-60,0])
ax.set_ylim([30,100])
```
