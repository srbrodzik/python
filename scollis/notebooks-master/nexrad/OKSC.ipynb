```python
#first we do some imports and check the version of Py-ART for consistency
import pyart
from matplotlib import pyplot as plt
import netCDF4
import numpy as np
from copy import deepcopy
import os
%matplotlib inline
print pyart.__version__
```

```python
#you can grab the data here: http://engineering.arm.gov/~collis/KAMX_20140417_1056
filename = '/data/radar/KFDR_20140421_0026'
radar = pyart.io.read(filename)
```

```python
display = pyart.graph.RadarMapDisplay(radar)
f = plt.figure(figsize = [17,4])
plt.subplot(1, 3, 1) 
display.plot_ppi_map('differential_reflectivity', max_lat = 35., min_lat =34, min_lon = -101., max_lon = -99,
                     vmin = -1, vmax = 4, lat_lines = np.arange(34,35,.2), lon_lines = np.arange(-101, -99, .5),
                     resolution = 'l')
plt.subplot(1, 3, 2) 
display.plot_ppi_map('reflectivity',   max_lat = 35., min_lat =34, min_lon = -101., max_lon = -99,
                     vmin = -8, vmax = 72, lat_lines = np.arange(34,35,.2), lon_lines = np.arange(-101, -99, .5),
                     resolution = 'l')
plt.subplot(1, 3, 3) 
display.plot_ppi_map('cross_correlation_ratio',  max_lat = 35., min_lat =34, min_lon = -101., max_lon = -99,
                     vmin = .7, vmax = 1., lat_lines = np.arange(34,35,.2), lon_lines = np.arange(-101, -99, .5),
                     resolution = 'l')

 
```

```python
smooth_zdr = deepcopy(radar.fields['differential_reflectivity'])
for i in range(smooth_zdr['data'].shape[0]):
    smooth_zdr['data'][i,:] = \
            pyart.correct.phase_proc.smooth_and_trim(smooth_zdr['data'][i,:], 8)

radar.fields.update({'differential_reflectivity_smooth': smooth_zdr})
```

```python
display = pyart.graph.RadarMapDisplay(radar)
f = plt.figure(figsize = [17,4])
plt.subplot(1, 3, 1) 
display.plot_ppi_map('differential_reflectivity_smooth', max_lat = 35., min_lat =34, min_lon = -100.3, max_lon = -99.25,
                     vmin = -1, vmax = 4, lat_lines = np.arange(34,35,.2), lon_lines = np.arange(-101.5, -99, .5),
                     resolution = 'l')
plt.subplot(1, 3, 2) 
display.plot_ppi_map('reflectivity',   max_lat = 35., min_lat =34, min_lon = -100.3, max_lon = -99.25,
                     vmin = -8, vmax = 72, lat_lines = np.arange(34,35,.2), lon_lines = np.arange(-101.5, -99, .5),
                     resolution = 'l')
plt.subplot(1, 3, 3) 
display.plot_ppi_map('cross_correlation_ratio',  max_lat = 35., min_lat =34, min_lon = -100.3, max_lon = -99.25,
                     vmin = .7, vmax = 1., lat_lines = np.arange(34,35,.2), lon_lines = np.arange(-101.5, -99, .5),
                     resolution = 'l')

 
```
