Mapping a whole file to a Cartesain can take time when often you just want a pretty image that slices the data.. To the end we can just create three 2D grids and use our plotting software to visualize them.. Great for quick investigation before doing full blown analysis (eg looking at data on the laptop!)

```python
import pyart
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import rc
from scipy.optimize import curve_fit
from copy import deepcopy
from time import time
%matplotlib inline
```

Read two NEXRAD files 

```python
radar = pyart.io.read('/data/radar/KATX_20140319_1402')
radar2 = pyart.io.read('/data/radar/KRTX_20140319_1401')
```

Plot a PPI near Seattle

```python
font = {'size' : 16}
rc('font', **font)
f = plt.figure(figsize=[15,8])
myd = pyart.graph.RadarMapDisplay(radar)
myd.plot_ppi_map('reflectivity', 1, vmin=-20, vmax=64, 
                 resolution = 'i',min_lon=-125, max_lon=-121, min_lat=46.3, max_lat=50,
                 lon_lines = np.arange(-125, -121, 1.), 
                 lat_lines = np.arange(45, 50, .5),
                 auto_range=False)
ax1 = plt.gca()
x0,x1 = ax1.get_xlim()
y0,y1 = ax1.get_ylim()
ax1.set_aspect((x1-x0)/(y1-y0))
```

Create three grid objects.. since we are not 3D we can go nice and high resolution in the vertical. 

```python
time_start = time()
grid_cappi = pyart.map.grid_from_radars(
    (radar,radar2),
    grid_shape=(1, 201, 201),
    grid_limits=((1500., 1500.), (-150000., 150000.), (-150000., 150000.)),
    fields=radar.fields.keys(), refl_field='reflectivity')
print 'Done cappi'
grid_yslice = pyart.map.grid_from_radars(
    (radar,radar2),
    grid_shape=(32, 1, 201),
    grid_limits=((0., 15000.), (0., 0.), (-150000., 150000.)),
    fields=radar.fields.keys(), refl_field='reflectivity')
print 'Done yslice'
grid_xslice = pyart.map.grid_from_radars(
    (radar,radar2),
    grid_shape=(32, 201, 1),
    grid_limits=((0., 15000.), (-150000., 150000.),(0., 0.)),
    fields=radar.fields.keys(), refl_field='reflectivity')
print 'Done xslace'
print 'Total time: ', time() - time_start 
```

so 77 Seconds, not blazing but not bad for on my laptop.. 
Now we use three display objects to create the "bird's eye" panel and the constant longitude and latitude slices. 

```python
def nice_map(grid_z, grid_y, grid_x, moment, lat = 36.6, 
             lon = -97.7, level = 0, vmin = -20, vmax = 64):
    # panel sizes
    map_panel_axes = [0.05, 0.05, .4, .80]
    x_cut_panel_axes = [0.55, 0.10, .4, .30]
    y_cut_panel_axes = [0.55, 0.50, .4, .30]
    colorbar_panel_axes = [0.05, 0.90, .4, .03]
    
    #min_lon=-157.1, max_lon=-156, min_lat=71.2, max_lat=71.6
    
    # parameters   
    fig = plt.figure(figsize=[15, 8])
    
    grid_display_z = pyart.graph.GridMapDisplay(grid_z)
    
    
    ax1 = fig.add_axes(map_panel_axes)
    grid_display_z.plot_basemap(resolution = 'i',
                              lon_lines = np.arange(-125, -121, 1.), 
                              lat_lines = np.arange(45, 50, .5))
    
    grid_display_z.plot_grid(moment, 
                           level=level, vmin=vmin, vmax=vmax)
    
    grid_display_z.plot_crosshairs(lon=lon, lat=lat)
    
    # colorbar
    cbax = fig.add_axes(colorbar_panel_axes)
    grid_display_z.plot_colorbar(cax=cbax)
    
    # panel 2, longitude slice.
    ax2 = fig.add_axes(x_cut_panel_axes)
    grid_display_x = pyart.graph.GridMapDisplay(grid_x)
    grid_display_x.plot_longitude_slice(moment, 
                                      vmin=vmin, vmax=vmax, lon=lon, lat=lat)
    ax2.set_xlabel('Distance from radar (km)')
    
    # panel 3, latitude slice
    ax3 = fig.add_axes(y_cut_panel_axes)
    grid_display_y = pyart.graph.GridMapDisplay(grid_y)
    grid_display_y.plot_latitude_slice(moment,
                                     vmin=vmin, vmax=vmax, lon=lon, lat=lat)
    
    x0,x1 = ax1.get_xlim()
    y0,y1 = ax1.get_ylim()
    ax1.set_aspect((x1-x0)/(y1-y0)-.15)

nice_map(grid_cappi, grid_yslice,grid_xslice, 'reflectivity', vmin = -20, vmax = 64, 
         lat = radar.latitude['data'], lon = radar.longitude['data'] )
```
