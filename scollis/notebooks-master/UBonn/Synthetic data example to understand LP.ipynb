```python
# For testing variable aspect ratio
def drop_ar(D_eq):
    if D_eq < 0.7:
        return 1.0;
    elif D_eq < 1.5:
        return 1.173 - 0.5165*D_eq + 0.4698*D_eq**2 - 0.1317*D_eq**3 - \
            8.5e-3*D_eq**4
    else:
        return 1.065 - 6.25e-2*D_eq - 3.99e-3*D_eq**2 + 7.66e-4*D_eq**3 - \
            4.095e-5*D_eq**4 
```

Lets make up some data, 100m range gates.

```python
range_from_radar = np.arange(0, 100.0*1000.0, 100.0)
```

Lets put a block of returns between 25 and 50km 
with a core from 35 to 40

```python
KDP = np.zeros(range_from_radar.shape)

start_index = range_from_radar.searchsorted(25.0*1000.0)
end_index = range_from_radar.searchsorted(50.0*1000.0)
KDP[start_index:end_index] = 4.0 #deg/km

start_index = range_from_radar.searchsorted(35.0*1000.0)
end_index = range_from_radar.searchsorted(40.0*1000.0)
KDP[start_index:end_index] = 6.0 #deg/km

PHIDP = np.cumsum(KDP)*(range_from_radar[1]- range_from_radar[0])/1000.0
```

```python
f = plt.figure(figsize = [10,5])
plt.plot(range_from_radar, PHIDP)
plt.ylabel('Propigation phase (deg/km)')
plt.twinx()
plt.plot(range_from_radar, KDP, '-r')
plt.ylabel('Specific differential phase (deg)')
plt.xlabel('Range (m)')
```
