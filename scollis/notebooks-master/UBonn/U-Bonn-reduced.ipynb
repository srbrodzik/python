```python
field_translation = {'PHIdpglatt': 'glatt_dp_phase_shift',  
                     'Kdpplot' : 'plot_diff_phase',
                     'PHIdpplot' : 'plot_dp_phase_shift',
                     'Zdrplot' : 'plot_diff_reflectivity',
                     'delta' : 'delta_phase_shift',
                     'Zh' : 'reflectivity_horizontal', 
                     'Zv' : 'reflectivity_vertical', 
                     'Zhfil' : 'corrected_reflectivity_horizontal', 
                     'Kdp' : 'diff_phase', 
                     'PHIdp' : 'dp_phase_shift',
                     'Zdr' : 'diff_reflectivity',
                     'Uh' : 'raw_reflectivity_horizontal', 
                     'Uv' : 'raw_reflectivity_vertical', 
                     'Vh' : 'mean_doppler_velocity', 
                     'Vv' : 'vertical_mean_doppler_velocity', 
                     'Rhv' : 'copol_coeff'}


ngates = len(mydataset.dimensions['range'])
nrays = len(mydataset.dimensions['azi'])
myradar = pyart.testing.make_empty_ppi_radar(ngates, nrays, 1)
for field in field_translation.keys(): 
    #print(field)
    field_dict = {'data' : mydataset.variables[field][:],
                  'units' : 'Unk',
                  'long_name' : mydataset.variables[field].long_name}
    myradar.add_field(field_translation[field], field_dict)   
myradar.range['data'] = np.linspace(0,ngates*100.0, ngates)

#Lon	7.071663° E
#Lat	50.73052° N
#Height ASL	99.5 m
myradar.latitude['data'] = 50.73052
myradar.longitude['data'] = 7.071663
myradar.altitude['data'] = 99.5
myradar.azimuth['data']= mydataset.variables['Azimut'][:]
myradar.metadata['instrument_name'] = 'UBonn X-Band'
```

Now we are going to run the stardard ARM processing on the phase data as showing in (Giangrande et al 2013). This requires us to set the FZL (going to set it to 2km) and the system phase and fill in a Normalized Coherent Power field which identifies first trip echoes. 

```python
#hard set the FZL
FZL = 2000.0
#hard set system phase
sysphase = -93.0
#create false NCP field 

ncp_dict = {'data' : np.ones(myradar.fields['copol_coeff']['data'].shape),
            'standard_name' : 'signal_quality_index',
            'long_name' : 'normalized_coherent_power'}

myradar.add_field('norm_coherent_power', ncp_dict)
```

Now fun the LP Processing code

```python
overide_sys_phase=False
nowrap=None
really_verbose=False
refl_field = 'reflectivity_horizontal'
ncp_field ='norm_coherent_power'
rhv_field = 'copol_coeff'
phidp_field = 'dp_phase_shift'
kdp_field = 'specific_differential_phase'
unf_field = 'unfolded_differential_phase'
offset = 0 
low_z=10.0
high_z=53.0
min_phidp=0.01
min_ncp=0.5
min_rhv=0.8
fzl=4000.0
sys_phase=0.0
self_const=60000.0
self_const_upper=60000.0

my_unf = pyart.correct.phase_proc.get_phidp_unf(myradar, ncp_lev=min_ncp, rhohv_lev=min_rhv,
                           debug=False, ncpts=2, doc=None,
                           sys_phase=sys_phase, nowrap=nowrap,
                           overide_sys_phase=overide_sys_phase,
                           refl_field=refl_field, ncp_field=ncp_field,
                           rhv_field=rhv_field, phidp_field=phidp_field)
my_new_ph = copy.deepcopy(myradar.fields[phidp_field])
my_unf[:, -1] = my_unf[:, -2]
my_new_ph['data'] = my_unf
myradar.fields.update({unf_field: my_new_ph})
```

```python
# prepare reflectivity field
refl = copy.deepcopy(myradar.fields[refl_field]['data']) + offset
is_low_z = (refl) < 20.0
is_high_z = (refl) > high_z
refl[np.where(is_high_z)] = high_z
refl[np.where(is_low_z)] = 0.0
z_mod = refl

refl = copy.deepcopy(myradar.fields[refl_field]['data']) + offset
is_low_z = (refl) < 20.0
is_high_z = (refl) >= 20.0
refl[np.where(is_high_z)] = 65.
refl[np.where(is_low_z)] = 1.0
z_mod_upper = refl
```

```python
phidp_mod = copy.deepcopy(myradar.fields[unf_field]['data'])
phidp_neg = phidp_mod < min_phidp
phidp_mod[np.where(phidp_neg)] = min_phidp
# process
proc_ph = copy.deepcopy(myradar.fields[phidp_field])
proc_ph['data'] = phidp_mod
St_Gorlv_differential_5pts = [-.2, -.1, 0, .1, .2]
```

Note det_process_range comes from <a href='https://github.com/ARM-DOE/pyart/blob/master/pyart/correct/phase_proc.py#L139'>here </a>
Ok, now is where we have to get clever. $\bf{A}$ is constant for the radar (the radar object pads gates with _Fill_Value, making the time-range array rectangular). But the B vector of the objective constraints is not as it contains reflectivity info. Now we want to use Scientific Python array based processing to take advantage of the optimized code backend.
The B vector is the negatibe and positive modified unfolded $\phi_{DP}$ and the self consistency constraint which uses converts $Z$ to $K_{DP}$ and integrates to give a $\phi_{DP}$ like field. 

$\bf{B}=\begin{bmatrix} -\bf{\phi_{DP}} \\\\ \bf{\phi_{DP}} \\\\ \bf{\eta} \end{bmatrix}$

Now we want to array-ize this effectively making $\bf{B}$ a matrix consiting of:

$\bf{B_{mat}}=\begin{bmatrix} -\bf{\phi_{DP}^{(1)}} & \cdots & -\bf{\phi_{DP}^{(n)}} \\\\ \bf{\phi_{DP}^{(1)}} & \cdots & \bf{\phi_{DP}^{(n)}} \\\\ \bf{\eta}^{(1)} & \cdots & \bf{\eta}^{(n)} \end{bmatrix}$

 where n is the number of rays. 
 
 Now for the self consistency constraint.. We are going want a gradient in $\phi_{DP}$ consistent with $Z_{dB}$ according to:

$\eta=\frac{(10^{0.1*Z_{dB}})^{0.914}}{w}$

where $w$ is the (inverse) weight, and is determined by trail and error, and in our case is set to 50,000 

```python

sweep = 0

end_gate, start_ray, end_ray = pyart.correct.phase_proc.det_process_range(
            myradar, sweep, FZL, doc=15)

start_gate = 0

A_Matrix = pyart.correct.phase_proc.construct_A_matrix(
            len(myradar.range['data'][start_gate:end_gate]),
            St_Gorlv_differential_5pts)

B_vectors =  pyart.correct.phase_proc.construct_B_vectors(
            phidp_mod[start_ray:end_ray, start_gate:end_gate],
            z_mod[start_ray:end_ray, start_gate:end_gate],
            St_Gorlv_differential_5pts, dweight=self_const)

B_vectors_upper =  pyart.correct.phase_proc.construct_B_vectors(
            phidp_mod[start_ray:end_ray, start_gate:end_gate],
            z_mod_upper[start_ray:end_ray, start_gate:end_gate],
            St_Gorlv_differential_5pts, dweight=self_const)


weights = np.ones(
            phidp_mod[start_ray:end_ray, start_gate:end_gate].shape)

nw = np.bmat([weights, np.zeros(weights.shape)])
print A_Matrix.shape
print B_vectors.shape
```

```python
n_gates = nw.shape[1]/2
n_rays = B_vectors.shape[0]
soln = np.zeros([n_rays, n_gates])
```

```python
model = CyLPModel()
G = np.matrix(A_Matrix)
h = CyLPArray(np.empty(B_vectors.shape[1]))
x = model.addVariable('x', G.shape[1])
model.addConstraint(G * x >= h)


#c = CyLPArray(np.empty(weights.shape[1]))
c = CyLPArray(np.squeeze(nw[0]))
model.objective = c * x

# import model in solver
s = CyClpSimplex(model)

az_desired = 220.
raynum = np.abs(myradar.azimuth['data'] - az_desired).argsort()[0]
print raynum, myradar.azimuth['data'][raynum]
# set new B_vector values for actual ray
s.setRowLowerArray(np.squeeze(np.asarray(B_vectors[raynum])))
s.setRowUpperArray(np.squeeze(np.asarray(B_vectors_upper[raynum])))
# set new weights (objectives) for actual ray
s.setObjectiveArray(np.squeeze(np.asarray(weights[raynum])))
# solve with dual method, it is faster
s.dual()
# extract primal solution
this_ray_soln = s.primalVariableSolution['x'][n_gates: 2 * n_gates]

# apply smoothing filter on a per scan basis
filt_soln = pyart.correct.phase_proc.smooth_and_trim(this_ray_soln, window_len=5, window='sg_smooth')
window_len = 35
sobel = 2. * np.arange(window_len)/(window_len - 1.0) - 1.0
sobel = sobel/(abs(sobel).sum())
sobel = sobel[::-1]
gate_spacing = (myradar.range['data'][1] - myradar.range['data'][0]) / 1000.
kdp = (scipy.ndimage.filters.convolve1d(filt_soln, sobel) /
           ((window_len / 3.0) * 2.0 * gate_spacing))
Delt = myradar.fields[unf_field]['data'][raynum,0:end_gate] - filt_soln
```

```python
f = plt.figure(figsize = [15,8])
plt.plot(myradar.range['data'][0:end_gate]/1000.0,
         filt_soln, 'r-', label = 'LP $\phi_{dp}$ solution, 5pt')
plt.plot(myradar.range['data'][0:end_gate]/1000.0,
         myradar.fields[unf_field]['data'][raynum,0:end_gate],'k-', label = 'Unfolded raw $\phi_{dp}$' )
plt.plot(myradar.range['data'][0:end_gate]/1000.0,
         myradar.fields['delta_phase_shift']['data'][raynum,0:end_gate],'k--', label = 'UBonn $\delta_{dp}$')
plt.plot(myradar.range['data'][0:end_gate]/1000.0,
         Delt,'r--', label = 'LP $\delta_{dp}$')

plt.xlabel('Distance (km)')
plt.ylabel('Phase shift (degrees)')
plt.gca().set_ylim([-10, 40])

plt.legend(loc = 4)
plt.twinx()
plt.plot(myradar.range['data'][0:end_gate]/1000.0,kdp, 'b-', label = 'LP $K_{dp}$ (sobel 35pt)')
plt.plot(myradar.range['data'][0:end_gate]/1000.0,
         myradar.fields[refl_field]['data'][raynum,0:end_gate]/10.0, 'y-', label = '$Z_e$/10')

plt.plot(myradar.range['data'][0:end_gate]/1000.0,
         myradar.fields['diff_reflectivity']['data'][raynum,0:end_gate], 'b--', label = '$Z_{DR}$')

plt.ylabel('$Z_{dr}$(dB), $Z_h/10$(dBZ), $K_{dp}$(deg/km)')
plt.legend(loc = 3)

ax = plt.gca()
ax.set_xlim([0,30])
ax.set_ylim([-5,10])
```
