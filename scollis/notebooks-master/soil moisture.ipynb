```python
soil = np.empty([0, 8])
watcont = np.empty([0, 8])
time = np.empty(0)
for i in range(len(swats)):
    try:
        myncf = netCDF4.Dataset('/data/swats/'+ swats[i])
        #print myncf.variables['soilwatpot_W'].shape
        #print i*6,(i+1)*6
        if myncf.variables['soilwatpot_W'].shape[1] == 8:
            soil = np.append(soil, myncf.variables['soilwatpot_W'][:], axis = 0)
            watcont = np.append(watcont, myncf.variables['watcont_W'][:], axis = 0)
            time = np.append(time, mdates.date2num(netCDF4.num2date(myncf.variables['time_offset'][:], 
                                                                myncf.variables['time_offset'].units)))
        else:
            data_shape = myncf.variables['soilwatpot_W'].shape
            pad = np.zeros([data_shape[0], 8 - data_shape[1]])
            soil = np.append(soil, np.append(myncf.variables['soilwatpot_W'][:], pad, axis = 1 ), axis = 0)
            watcont = np.append(watcont, np.append(myncf.variables['watcont_W'][:], pad, axis = 1), axis = 0)
            time = np.append(time, mdates.date2num(netCDF4.num2date(myncf.variables['time_offset'][:], 
                                                                myncf.variables['time_offset'].units)))
            myncf.close()
    except KeyError:
        print "Failed ", swats[i]
```
