```python
import netCDF4
from matplotlib import pyplot as plt
import pyproj
import numpy as np
from mpl_toolkits.basemap import Basemap
import urllib2
%matplotlib inline
```

```python
dir_string = 'http://motherlode.ucar.edu:8080/thredds/dodsC/satellite/IR/EAST-CONUS_4km/current/'
dirlisting = urllib2.urlopen(dir_string)
all_lines = dirlisting.read().splitlines()
has_gini = []
for line in all_lines:
    if '.gini' in line:
        #print line
        happy_place = line.find('EAST-CONUS_4km_IR_')
        end_place = line.find('.gini')
        my_upper_name = line[happy_place:end_place]+'.gini'
        url = dir_string + my_upper_name
        has_gini.append(url)
```

```python
print has_gini[0]
```

```python
data=netCDF4.Dataset(has_gini[0])
```

```python
vals=data.variables['IR'][:]*-1.0
swap_me = np.where(vals < 0)
vals[swap_me] = vals[swap_me] + 128.0*2.0

print vals.shape
xg, yg = np.meshgrid(data.variables['x'][:]*1000.0, data.variables['y'][::-1]*1000.0)

 
pnyc = pyproj.Proj(proj = 'lcc',datum = 'NAD83', 
                   lat_0 = data.variables['LambertConformal'].latitude_of_projection_origin,
                   lon_0 = data.variables['LambertConformal'].longitude_of_central_meridian,
                   x_0 = 0.0,
                   y_0 = 0.0)

lon, lat = pnyc(xg, yg, inverse = True)

min_lon = lon.min(); max_lon = lon.max()
min_lat = lat.min(); max_lat = lat.max()

box=[-100, -80, 40, 50]

m = Basemap(llcrnrlon=box[0],llcrnrlat=box[2],
            urcrnrlon=box[1],urcrnrlat=box[3],
            projection='lcc',area_thresh=10000,
            resolution='h', 
            lat_0 = data.variables['LambertConformal'].latitude_of_projection_origin,
            lon_0 = data.variables['LambertConformal'].longitude_of_central_meridian)

x, y = m(lon, lat)
# create figure.
fig=plt.figure(figsize=(10,9))
ax = fig.add_axes([0.01,0.05,0.9,0.80]) 
m.drawparallels(np.linspace(30,50, 11) ,labels=[1,1,0,0])
m.drawmeridians(np.linspace(-100, -80,11),labels=[0,0,0,1]) 
m.pcolormesh(x, y , vals[0,::-1], cmap=plt.get_cmap('binary'))
xr, yr = m([box[0], box[1]], [box[2], box[3]])
plt.xlim(xr)
plt.ylim(yr)

m.drawcoastlines(linewidth=1.25)
m.drawstates()
plt.colorbar()
```
