```python
def extract_rhi(radar, field, azimuths):
    """
    Extract a len(angles) sweep RHI from PPI and positions angles. 

    Parameters
    ----------
    radar : A PPI radar object 
        Radar object to extract RHIs from.
    angles : list or numpy array
        Angles to extract RHIs at
    
    Returns
    -------
    rhi_array : Numpy masked array
        Array of shape [len(azimuths)*n_elevations, ppi_shape[1]]
        containing an RHI composed of the rays from the PPI
 
    """

    ppi_shape = radar.fields[radar.fields.keys()[0]]['data'].shape
    n_elevations = len(radar.sweep_number['data'])
    rhi_array = np.ma.zeros([len(azimuths)*n_elevations, ppi_shape[1]])
    azimuth_array = np.zeros(len(azimuths)*n_elevations)
    elevation_array = np.zeros(len(azimuths)*n_elevations)
    
    az_number = 0
    for azimuth in azimuths:
        for sweep in radar.sweep_number['data']:
            i_s = radar.sweep_start_ray_index['data'][sweep]
            i_e = radar.sweep_end_ray_index['data'][sweep]
            sweep_azimuths = radar.azimuth['data'][i_s:i_e]
            target_index = abs(sweep_azimuths - azimuth).argsort()[0]
            rhi_array[az_number*n_elevations + sweep, :] =\
                  radar.fields[field]['data'][i_s + target_index, :]
            azimuth_array[az_number*n_elevations + sweep] = \
                        radar.azimuth['data'][i_s + target_index]
            elevation_array[az_number*n_elevations+ sweep] = \
                        radar.elevation['data'][i_s + target_index]
        az_number = az_number + 1
    return rhi_array, azimuth_array, elevation_array
```

```python
def rhi_from_ppi(radar, angles, force = False):
    """
    Extract a len(angles) sweep RHI from PPI and positions angles. 

    Parameters
    ----------
    radar : A PPI radar object 
        Radar object to extract RHIs from.
    angles : list or numpy array
        Angles to extract RHIs at
    
    Returns
    -------
    radar : Radar
        Radar object with no fields, other parameters are set to default
        values.

    """
    #first do checks on the radar object to ensure it is a PPI
    
    if radar.scan_type != "ppi" and not(force):
        #this is not a PPI throw an exception and return
        
        raise TypeError('Expected PPI Volume')
    
    number_of_rhis = len(angles)
    n_elevations = len(radar.sweep_mode['data'])
    number_of_rays = number_of_rhis*n_elevations
    print number_of_rays
    synthetic_scan_type = 'rhi'
    #create some synthetic time
    
    synthetic_time = deepcopy(radar.time)
    synthetic_time['data'] = np.linspace(radar.time['data'][0],
                                         radar.time['data'][-1],
                                         number_of_rays)
    
    #grab the instrument parameters and trim them back
    
    synthetic_parameters = deepcopy(radar.instrument_parameters)
    
    for instr_param in synthetic_parameters.keys():
        synthetic_parameters[instr_param]['data'] = \
                  synthetic_parameters[instr_param]['data'][0:number_of_rhis]
    
    #grab and modify sweep information
    
    synthetic_sweep_number = deepcopy(radar.sweep_number)
    synthetic_sweep_number['data'] = synthetic_sweep_number['data'][0:number_of_rhis]
    
    synthetic_sweep_mode = deepcopy(radar.sweep_mode)
    synthetic_sweep_mode['data'] = np.array(number_of_rhis*['rhi'])
    
    synthetic_fixed_angle = deepcopy(radar.fixed_angle)
    synthetic_fixed_angle['data'] = np.array(angles)
    
    synthetic_sweep_start_ray_index = deepcopy(radar.sweep_start_ray_index)
    synthetic_sweep_start_ray_index['data'] = np.arange(0, 
                                                  number_of_rays - n_elevations+1, 
                                                  n_elevations)
    
    synthetic_sweep_end_ray_index = deepcopy(radar.sweep_end_ray_index)
    synthetic_sweep_end_ray_index['data'] = np.arange( n_elevations, 
                                                  number_of_rays+1, 
                                                  n_elevations)
    synthetic_fields = {}
    
    for ppi_field in radar.fields.keys():
        this_syn_rhi, azm, ele = extract_rhi(radar, ppi_field, angles)
        this_syn_field = deepcopy(radar.fields[ppi_field])
        this_syn_field['data'] = this_syn_rhi
        synthetic_fields.update({ppi_field:this_syn_field})
    
    syn_azmths = deepcopy(radar.azimuth)
    syn_azmths['data'] = azm
    syn_ele = deepcopy(radar.elevation)
    syn_ele['data'] = ele
    
    synthetic_radar = pyart.io.radar.Radar(synthetic_time, deepcopy(radar.range), synthetic_fields,
                                           deepcopy(radar.metadata), synthetic_scan_type, 
                                           deepcopy(radar.latitude), deepcopy(radar.longitude), 
                                           deepcopy(radar.altitude), synthetic_sweep_number,
                                           synthetic_sweep_mode, synthetic_fixed_angle, 
                                           synthetic_sweep_start_ray_index,
                                           synthetic_sweep_end_ray_index, syn_azmths, syn_ele,
                                           instrument_parameters=synthetic_parameters)
    
    return synthetic_radar
```

```python
for ray in rhi_radar.sweep_number['data']:
    # Lets plot some RHIs
    
    # So first we are going to create a list of pairs of strings..
    # these are the two fields to be plotted per line, so in the end we
    # have 4 rows and two columns
    
    field_pairs = [['reflectivity', 'velocity'],
                   ['differential_reflectivity', 'differential_phase'],
                   ['differential_reflectivity_smooth', 'cross_correlation_ratio'],
                   ['specific_differential_phase', 'spectrum_width']]
    
    # so this is a list of yuples. (two-tuples to be exact)
    # just like the above!
    # these are the value ranges for each of the plots.. please feel free to experiment!
    
    range_pairs= [[(-8, 72), (-32.0, 32.0)],
                  [(-1, 4.0), (-80, 160.0)],
                  [(-1, 4.0), (.7, 1.0)],
                  [(-2.0, 6), (0., 4.0)]]
    
    # just like above, create a new display
    
    display = pyart.graph.RadarDisplay(rhi_radar)
    
    # Time to loop folks! Note the way Python deals with
    # flow control. Indentation.. Deal with it and I promise
    # in the end you will love it. 
    
    # looping is done over a list.. the range() function creates 
    # a list from 0 to len(field_pairs)-1
    
    for i in range(len(field_pairs)):
        # fields to plot and ranges
        fields_to_plot = field_pairs[i]  # this is now just a pair of strings
        ranges = range_pairs[i]  # this is now a pair of two-tuples
        
        # plot the data
        nplots = len(fields_to_plot)  # This will be 2
        
        #create a matplotlib figure object
        plt.figure(figsize=[7 * nplots, 3])  # you can play with this! 
        
        # plot each field, so here we loop over the two fields per row
        for plot_num in range(nplots): 
            field = fields_to_plot[plot_num]  # grab the field to be plotted for this panel
            vmin, vmax = ranges[plot_num]  # grab the ranges from the two-Tuple
            
            # The below command is a matplotlib command. This creates
            # subplots (Rows, Columns, and current plot)
            # So on the first execution it is a 1 row 2 column at the 1st position
            # and on the second at the 2nd position 
            plt.subplot(1, nplots, plot_num + 1) 
            display.plot_rhi(field, ray, vmin=vmin, vmax=vmax, title_flag=False)  # and you know this! 
            display.set_limits(ylim=[0, 12], xlim = [0,45])  # feel free to play and to add a xlim below!
        
        # set the figure title
        radar_name = display.radar_name  # grab the name of the radar from the display
        time_text = ' ' + display.time_begin.isoformat() + 'Z ' # the display obj helps you! 
        azimuth = rhi_radar.fixed_angle['data'][ray] 
        title = 'RHI ' + radar_name + time_text + 'Azimuth %.2f' % (azimuth)
        plt.suptitle(title)
```
