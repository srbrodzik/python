```python
c_only = pyart.io.read_grid('/data/scipy2014/c_mesh_QC_highres.nc')
```

```python
def nice_three_panel(grid, var, level, vmin, vmax, lat = 36.57861, lon = -97.763611):
    display = pyart.graph.GridMapDisplay(grid)
    # create the figure
    font = {'size': 16}
    rc('font', **font)
    fig = plt.figure(figsize=[15, 8])

    # panel sizes
    map_panel_axes = [0.05, 0.05, .4, .80]
    x_cut_panel_axes = [0.55, 0.10, .4, .30]
    y_cut_panel_axes = [0.55, 0.50, .4, .30]
    colorbar_panel_axes = [0.05, 0.90, .4, .03]

    # panel 1, basemap, radar reflectivity a
    ax1 = fig.add_axes(map_panel_axes)
    display.plot_basemap(lat_lines = np.arange(30,37,.2),
                         lon_lines = np.arange(-98, -96, .25))
    display.plot_grid(var, level=level, vmin=vmin, vmax=vmax)
    display.plot_crosshairs(lon=lon, lat=lat)

    # colorbar
    cbax = fig.add_axes(colorbar_panel_axes)
    display.plot_colorbar(cax=cbax)

    # panel 2, longitude slice.
    ax2 = fig.add_axes(x_cut_panel_axes)
    display.plot_longitude_slice(var, lon=lon, lat=lat, vmin=vmin, vmax=vmax)
    ax2.set_xlabel('Distance from SGP CF (km)')

    # panel 3, latitude slice
    ax3 = fig.add_axes(y_cut_panel_axes)
    display.plot_latitude_slice(var, lon=lon, lat=lat, vmin=vmin, vmax=vmax)

    # add a title
    slc_height = grid.axes['z_disp']['data'][level]
    dts = num2date(grid.axes['time']['data'],grid.axes['time']['units'])
    datestr = dts[0].strftime('%H:%M Z on %Y-%m-%d')
    title = 'Sliced at ' + str(slc_height) + ' meters at ' + datestr
    fig.text(0.5, 0.9, title)
```

```python
nice_three_panel(c_only, 'rain_rate_A', 1, 0, 80, lat = 36.5, lon = -97.65)
```

```python
def area_anal(pixel_area, rr_x, rain_rates):
    A_rainrate=np.zeros(rr_x.shape)
    N_rainrate=np.zeros(rr_x.shape)
    Rm_rainrate=np.zeros(rr_x.shape)
    my_shape=rain_rates.shape
    for i in range(len(rr_x)):
        b_fld=np.zeros(my_shape)
        b_fld[rain_rates > rr_x[i]]=1.0
        regions, N_rainrate[i] = ndimage.label(b_fld)
        try:
            A_rainrate[i]=len(np.where(regions > 0.5)[0])*pixel_area
            Rm_rainrate[i]=rain_rates[np.where(regions > 0.5)].mean()
        except IndexError:
            A_rainrate[i]=0.0
            Rm_rainrate[i]=0.0
    return N_rainrate, A_rainrate, Rm_rainrate
        
        
        
```

```python
rain_c = c_only.fields['rain_rate_A']['data'][1]
rr_x=np.linspace(0, 80.0, 100.0)
c_N,c_A,c_Rm=area_anal(.5 * .5, rr_x, rain_c)
```

```python
f=plt.figure(figsize=[8,8])
ax1=plt.subplot(3,1,1)
ax1.plot(rr_x, c_N)
ax2=plt.subplot(3,1,2)
a_radar  = 90.*100.
print a_radar
ax2.plot(rr_x, c_A/a_radar)
ax3=plt.subplot(3,1,3)
ax3.plot(rr_x, c_Rm)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.0, hspace=0.0)
ax1.get_xaxis().set_visible(False)
ax2.get_xaxis().set_visible(False)
ax2.yaxis.set_major_locator(plt.MaxNLocator(prune='upper'))
ax3.yaxis.set_major_locator(plt.MaxNLocator(prune='upper'))
ax1.set_ylabel('Number of regions')
ax2.set_ylabel('Total area of regions \n (% of radar coverage)')
ax3.set_ylabel(r'Mean rain rate (mmhr$^{-1}$)')
ax3.set_xlabel(r'Rain rate threshold (mmhr$^{-1}$)')

```

```python
rain_x = x_only.fields['rain_rate_A']['data'][1]
rr_x=np.linspace(0, 80.0, 100.0)
x_N,x_A,x_Rm=area_anal(.5 * .5, rr_x, rain_x)
```

```python
t=netCDF4.num2date(c_only.axes['time']['data'][0], c_only.axes['time']['units'],'gregorian')

xyt=[c_only.axes['x_disp']['data']/1000.0,c_only.axes['y_disp']['data']/1000.0, [t]]

this_rain=rain_c
b_fld=np.zeros(this_rain.shape)
b_fld[this_rain > 5.0]=1.0
regions_5, N_5 = ndimage.label(b_fld)
b_fld=np.zeros(this_rain.shape)
b_fld[this_rain > 20.0]=1.0
regions_40, N_40 = ndimage.label(b_fld)


f=plt.figure(figsize=[5,10])
ax1=plt.subplot(3,1,1)
pc1=ax1.pcolormesh(xyt[0], xyt[1], this_rain, vmin=0, vmax=120)
cb1=plt.colorbar(mappable=pc1)
cb1.set_label(r'Rain rate (mmhr$^{-1}$)')
ax2=plt.subplot(3,1,2)
pc2=ax2.pcolormesh(xyt[0], xyt[1], regions_5)
cb2=plt.colorbar(mappable=pc2)
cb2.set_label(r'n$_{reg}$ R$_\mathrm{thresh}$=5 mmhr$^{-1}$')
ax3=plt.subplot(3,1,3)
pc3=ax3.pcolormesh(xyt[0], xyt[1], regions_40)
cb3=plt.colorbar(mappable=pc3)
cb3.set_label(r'n$_{reg}$ R$_{\mathrm{thresh}}$=20 mmhr$^{-1}$')

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.0, hspace=0.0)
ax1.get_xaxis().set_visible(False)
ax2.get_xaxis().set_visible(False)
ax2.yaxis.set_major_locator(plt.MaxNLocator(prune='upper'))
ax3.yaxis.set_major_locator(plt.MaxNLocator(prune='upper'))
ax1.set_ylabel('N/S displacement (km)')
ax2.set_ylabel('N/S displacement (km)')
ax3.set_ylabel('N/S displacement (km)')
ax3.set_xlabel('E/W displacement (km)')
plt.savefig('/data/simple_labels.png', dpi=250)
```
