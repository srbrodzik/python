```python
base_name = "/data/scipy2014/"
xnw_name = base_name + "XNW110520064058.RAW7HE3"
xsw_name = base_name + "XSW110520064208.RAW7HE5"
xse_name = base_name + "XSE110520064117.RAW7HE4"
c_name = base_name + "064739.mdv"
xnw_radar = pyart.io.read(xnw_name)
xsw_radar = pyart.io.read(xsw_name)
xse_radar = pyart.io.read(xse_name)
c_radar = pyart.io.read(c_name)
```

Processing C-SAPR

```python
nw_offset = 0.0
se_offset = 0.0
sw_offset = 0.0
c_offset = - 3.42
```

```python
# Perform a Linear programming phase processing to get the conditionally
# fitted phase profile and sobel filtered KDP field
gates = c_radar.range['data'][1] - c_radar.range['data'][0]
rge = 20.0 * 1000.0
ng = rge / gates

reproc_phase, sob_kdp = pyart.correct.phase_proc.phase_proc_lp(c_radar, c_offset, debug=True,nowrap=ng)
c_radar.fields.update({'recalculated_diff_phase': sob_kdp,
                         'proc_dp_phase_shift': reproc_phase})
```

```python
spec_at, cor_z = pyart.correct.attenuation.calculate_attenuation(c_radar, c_offset, 
                                                                 debug=True, ncp_min=0.4)
c_radar.fields.update({'specific_attenuation': spec_at})
c_radar.fields.update({'corrected_reflectivity': cor_z})
```

```python
R = 300.0 * (c_radar.fields['specific_attenuation']['data']) ** 0.89
rainrate = copy.deepcopy(c_radar.fields['unfolded_differential_phase'])
rainrate['data'] = np.ma.masked_where(np.logical_or((c_radar.fields['normalized_coherent_power']['data'] < 0.4),
        (c_radar.fields['cross_correlation_ratio']['data'] < 0.8)), R)
rainrate['valid_min'] = 0.0
rainrate['valid_max'] = 400.0
rainrate['standard_name'] = 'rainfall_rate'
rainrate['long_name'] = 'rainfall_rate'
rainrate['least_significant_digit'] = 1
rainrate['units'] = 'mm/hr'
c_radar.fields.update({'rain_rate_A': rainrate})
mask = c_radar.fields['corrected_reflectivity']['data'].mask
c_radar.fields['rain_rate_A']['data'][np.where(mask)]=0.0
c_radar.fields['rain_rate_A'].update({'comment':'Rain rate calculated from specific_attenuation, R=300.0*specific_attenuation**0.89, note R=0.0 where norm coherent power < 0.4 or rhohv < 0.8'})
```

```python
fields_to_plot = ['specific_attenuation', 'rain_rate_A']
ranges = [(0, 0.5), (0, 100)]
display = pyart.graph.RadarMapDisplay(c_radar)

nplots = len(fields_to_plot)
plt.figure(figsize=[7 * nplots, 4])
for plot_num in xrange(nplots):
    field = fields_to_plot[plot_num]
    vmin, vmax = ranges[plot_num]
    plt.subplot(1, nplots, plot_num + 1)
    display.plot_ppi_map(field, 0, vmin=vmin, vmax=vmax, lat_lines = np.arange(40,60,.2),
                         lon_lines =  np.arange(0,10,.4), resolution = 'l')
    display.basemap.drawrivers()
    display.basemap.drawcountries()
    display.plot_range_rings([20,40])
    #display.set_limits(ylim=[-100, 100], xlim=[-100,100])
```

```python
pyart.io.write_cfradial('/data/scipy2014/corrected_csapr.nc', c_radar)
```
