So this has bought these libraries into the namespace, lets do a very simple example that we all do from day to day, fitting a straight line:

```python
help(numpy.linspace)
```

```python
#adapted from http://www.scipy.org/Cookbook/LinearRegression
npts=50
my_x=numpy.linspace(-5.,5.,npts)
a=0.8; b=-4.
my_y=scipy.polyval([a,b], my_x)+scipy.randn(npts)
```

The above code creates a linear array from -5 to 5 (note the "." after the number.. this automatically makes it a float (rather than having to cast it) with npts (50). This object is a numpy "array". We then use scipy's polyval function to generate a 1st order (ie two coefficents $y=b+ax$) polynomail and we then use scipy's randn to add randomly distributed noise (adding two, same length, arrays together). Now lets plot it using matplotlib (pylab) to see what we have:

```python
f=pylab.figure(figsize=[10,5])
pylab.plot(my_x, my_y, 'ro')
pylab.ylabel('Temperature (c)')
pylab.xlabel('Time (Hours)')
```

So the first line generates a canvus, f. The calling variable, a **keyword** variable, sets the geometry. The second line plots the data using red circles. The string to symbol mapping is, unsuprisingly, the same as in Matlab. 

This generates an object in the samespace wich is the current axis.. you can always inspect the axis using pylab.gca() all commands that do not specify a plot to work on work on the current axis, such as the next two lines (above) which add labels to the axis. One very important command in Python is the help function.. eg:

```python
help(scipy.polyfit)
```

Ok lets put it all together to make a function that can plot our KAZR time height data:

```python
name='steve'
pos={'scott':'left', 'steve':'right'}
print pos[name]
```

```python
def plot_kazr(kncf, var, rng):
    rnges={'mean_doppler_velocity_copol':[float(kncf.nyquist_velocity.split(' ')[0])*-1.0, float(kncf.nyquist_velocity.split(' ')[0])],
           'reflectivity_copol':[-45,25],
           'reflectivity_xpol':[-45,25]}
    print rng
    f=pylab.figure(figsize=[10,5])
    pc=pylab.pcolormesh(kncf.variables['time'][rng[0]:rng[1]],  
        kncf.variables['range'][:], 
        kncf.variables[var][:][rng[0]:rng[1],:].transpose(), 
        vmin=rnges[var][0], vmax=rnges[var][1])
    pylab.xlabel(kncf.variables['time'].long_name+ 
        ' ('+ kncf.variables['time'].units+')')
    pylab.ylabel(kncf.variables['range'].long_name+ 
        ' ('+ kncf.variables['range'].units+')')
    cb=pylab.colorbar()
    cb.set_label(kncf.variables[var].long_name+ 
        ' ('+ kncf.variables[var].units+')')
```

<h2>Custom Classes</h2>
Cool eh? But not very OO.. Making the function into a class not only cleans things up nicely it gives the project much more extensibility, ie making it easier for users to add functionality... Lets do a basic hello world class first.. Now as before, objects have methods and data fields. When you define your own class there are two special methods you can define.. The __call__ and init method... Oh and then there is self... self is, well, the object itself!  

```python
class bob:
    def __init__(self, lunch):
        self.lunchbox=lunch
    def __call__(self, time):
        print "Eat ", self.lunchbox, " at ", time
        return "empty lunchbox"
    def add(self, item):
        self.lunchbox=self.lunchbox+ " and "+ item
```

```python
I=bob('fruit')
print dir(I)
print I.lunchbox
I.add('juice')
print I.lunchbox
I.add('steak')
a=I('10 am')
print a
```

```python
a=numpy.linspace(0,100,100)
print a.mean()
a()
```

So __init__ method is called when we first create an instance of bob. When this is called I is assigned to an instance of bob and I.lunchbox is assigned to the first positional arguement. The add method is self explanitory, we just add more strings to I.lunchbox. the __call__ method is invoked when we actually call I(). 

Ok now lets actually build something usefull. So here we want to use our __init__ method of our class to actually populate our instance with data that will be used when we invoke the __Call__ method to generate our plot... 

```python
class kazr_plotter:
    def __init__(self, kazr_nc):
        self.data=kazr_nc.variables
        self.time=kazr_nc.variables['time']
        self.range=kazr_nc.variables['range']
        self.value_ranges={'mean_doppler_velocity_copol':[float(kazr_nc.nyquist_velocity.split(' ')[0])*-1.0, float(kazr_nc.nyquist_velocity.split(' ')[0])],
           'reflectivity_copol':[-45,25],
           'reflectivity_xpol':[-45,25]}
        self.xlabel=self.time.long_name+ ' ('+ self.time.units+')'
        self.ylabel=self.range.long_name+ ' ('+ self.range.units+')'
    def __call__(self, variable, window):
        myfig=pylab.figure()
        self.pc=pylab.pcolormesh(self.time[window[0]:window[1]],  
        self.range[:]/1000.0, 
        self.data[variable][:][window[0]:window[1],:].transpose(), 
        vmin=self.value_ranges[variable][0], 
        vmax=self.value_ranges[variable][1])
        pylab.xlabel(self.xlabel)
        pylab.ylabel(self.ylabel)
        self.cb=pylab.colorbar()
        self.cb.set_label(self.data[variable].long_name+ 
        ' ('+ self.data[variable].units+')')
        
```

Ok.. this has all been fun, but lets actually do some analysis! We are going to pimp our class by adding in some super simple Contoured by Frequency Altitude Diagrams (CFAD), see (Yuter and Houze, 1995). Esssentially we are going to loop through our data, level by level and compute histograms of quantities... We wil conditionally sample above the radar minimum deteactable signal to ensure we are not summing incoherent signals. 

```python
class kazr_plotter2:
    def make_cfad(self, variable, window):
        nlevels=self.data[variable].shape[1]
        cfad=numpy.zeros([nlevels, 99], dtype=float)
        for i in range(nlevels):
            this_lev=self.data[variable][i, window[0]:window[1]].flatten()
            this_z=self.data['reflectivity_copol'][i, window[0]:window[1]].flatten()
            am_bigger_than_m25=numpy.where(this_z > -25.0)[0]
            if len(am_bigger_than_m25) > 5:
                filtered_this_lev=this_lev[am_bigger_than_m25]
                n, bins, patches = pylab.hist(filtered_this_lev, numpy.linspace(self.value_ranges[variable][0], self.value_ranges[variable][1], 100), normed=1, visible=False)
                cfad[i,:]=n
        return cfad, bins
    def __init__(self, kazr_nc):
        self.data=kazr_nc.variables
        self.time=kazr_nc.variables['time']
        self.range=kazr_nc.variables['range']
        self.value_ranges={'mean_doppler_velocity_copol':[float(kazr_nc.nyquist_velocity.split(' ')[0])*-1.0, float(kazr_nc.nyquist_velocity.split(' ')[0])],
           'reflectivity_copol':[-45,25],
           'reflectivity_xpol':[-45,25]}
        self.xlabel=self.time.long_name+ ' ('+ self.time.units+')'
        self.ylabel=self.range.long_name+ ' ('+ self.range.units+')'
    def __call__(self, variable, window, typ):
        if typ=='TH':
            figme=pylab.figure()
            self.pc=pylab.pcolormesh(self.time[window[0]:window[1]],  
            self.range[:]/1000.0, 
            self.data[variable][:][window[0]:window[1],:].transpose(), 
            vmin=self.value_ranges[variable][0], vmax=self.value_ranges[variable][1])
            pylab.xlabel(self.xlabel)
            pylab.ylabel(self.ylabel)
            self.cb=pylab.colorbar()
            self.cb.set_label(self.data[variable].long_name+ 
            ' ('+ self.data[variable].units+')')
        elif typ=='CF':
            self.mycfad, self.bins=self.make_cfad(variable, window)
            figme=pylab.figure()
            print "Plotting"
            self.pc=pylab.pcolormesh(self.bins, self.range[:], self.mycfad*100.0*(self.bins[1]-self.bins[0]), vmin=0, vmax=15)
            pylab.ylabel(self.ylabel)
            pylab.xlabel(self.data[variable].long_name+ 
            ' ('+ self.data[variable].units+')')
            self.cb=pylab.colorbar()
            self.cb.set_label('Percentage of Gates')
 
```

```python
my_plotter=kazr_plotter2(my_kazr)
```

```python
my_plotter('mean_doppler_velocity_copol', [0,1000], 'CF')
```

```python
my_plotter.value_ranges={'reflectivity_copol':[-25,20]}
my_plotter('reflectivity_copol', [0,500], 'CF')
```
