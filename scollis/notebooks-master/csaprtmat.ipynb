For scattering calculations we assume a Gamma DSD:

$N(D)=N_wf(\mu)(\frac{D}{D_0})e^{-\Lambda{}D}$

where:
$f(\mu)=\frac{6}{3.67^4}\frac{(3.67+\mu)^{\mu+4}}{\Gamma{}(\mu + 4)}$

$\Lambda{}=\frac{3.67 + \mu}{D_0}$

$N_0=N_wf(\mu)D_0^{-\mu}$

So first, instead of using impact parameter lets use something more physical.. Lets hold $N_0$ and $D_0$ constant for each curve which means $N_w$ varies as a function of $D_0$..

```python
mu=4.0
f_mu=(6.0/(3.67**4))*((3.67 + mu)**(mu+4.0))/gamma(mu+4.0)
print f_mu
n_d0=50
n_N0=7
D0=np.linspace(0.1, 3.0, n_d0)
N0=1*10**np.linspace(3,6.0,n_N0)#np.array([1e4, 5e4, 1e5, 5e5])
d0Nw=np.zeros([n_d0, n_N0])
for i in range(n_N0):
    d0Nw[:,i]=N0[i]/(f_mu*(D0**(-mu)))

```

```python
# For testing variable aspect ratio
def drop_ar(D_eq):
    if D_eq < 0.7:
        return 1.0;
    elif D_eq < 1.5:
        return 1.173 - 0.5165*D_eq + 0.4698*D_eq**2 - 0.1317*D_eq**3 - \
            8.5e-3*D_eq**4
    else:
        return 1.065 - 6.25e-2*D_eq - 3.99e-3*D_eq**2 + 7.66e-4*D_eq**3 - \
            4.095e-5*D_eq**4 
```

```python
zdr=np.zeros([n_d0,n_N0])
z=np.zeros([n_d0,n_N0])
kdp=np.zeros([n_d0,n_N0])
A=np.zeros([n_d0,n_N0])
for i in range(n_d0):
    for j in range(n_N0):
        tm = TMatrixPSD(lam=tmatrix_aux.wl_C, m=refractive.m_w_10C[tmatrix_aux.wl_C])
        tm.psd_eps_func = lambda D: 1.0/drop_ar(D)
        tm.D_max = 10.0
        tm.or_pdf = orientation.gaussian_pdf(20.0)
        tm.orient = orientation.orient_averaged_fixed
        tm.geometries = (tmatrix_aux.geom_horiz_back, tmatrix_aux.geom_horiz_forw)
        tm.init_scatter_table()
        tm.psd = GammaPSD(D0=D0[i], Nw=d0Nw[i,j], mu=mu)
        zdr[i,j]=radar.Zdr(tm)
        z[i,j]=radar.refl(tm)
        tm.set_geometry(tmatrix_aux.geom_horiz_forw)
        kdp[i,j]=radar.Kdp(tm)
        A[i,j]=radar.Ai(tm)

```

```python
def scatter_off_2dvd(d_diameters, d_densities):
    mypds = interpolate.interp1d(d_diameters,d_densities, bounds_error=False, fill_value=0.0)
    tm = TMatrixPSD(lam=tmatrix_aux.wl_C, m=refractive.m_w_10C[tmatrix_aux.wl_C])
    tm.psd_eps_func = lambda D: 1.0/drop_ar(D)
    tm.D_max = 10.0
    tm.or_pdf = orientation.gaussian_pdf(20.0)
    tm.orient = orientation.orient_averaged_fixed
    tm.geometries = (tmatrix_aux.geom_horiz_back, tmatrix_aux.geom_horiz_forw)
    tm.init_scatter_table()
    tm.psd = mypds#GammaPSD(D0=D0[i], Nw=Nw[j], mu=4)
    zdr=radar.Zdr(tm)
    z=radar.refl(tm)
    tm.set_geometry(tmatrix_aux.geom_horiz_forw)
    kdp=radar.Kdp(tm)
    A=radar.Ai(tm)
    return z,zdr,kdp,A
        
        
```
