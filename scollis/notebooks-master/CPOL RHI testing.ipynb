```python
import pyart
import matplotlib.pyplot as plt
import os
import netCDF4
%matplotlib inline


def nukedir(dir):
    if dir[-1] == os.sep: dir = dir[:-1]
    files = os.listdir(dir)
    for file in files:
        if file == '.' or file == '..': continue
        path = dir + os.sep + file
        if os.path.isdir(path):
            nukedir(path)
        else:
            os.unlink(path)
    os.rmdir(dir)


def dt_to_dict(dt, **kwargs):
        pref = kwargs.get('pref', '')
        return dict([(pref+key, getattr(dt, key)) for key in
                    ['year', 'month', 'day', 'hour', 'minute', 'second']])

```

```python

temp_output = '/data/temp_cf/'
nukedir(temp_output)
in_file = '/data/lassen/Gunn_pt_20060122171844_RHI.lassen'
rsl_exec = '/Users/scollis/anaconda/bin/any_to_uf'
radx_exec = '/usr/local/bin/RadxConvert'
param_file = '/data/cpol_to_cf.params'
temp_file = '/data/tempuf.uf'
temp_output = '/data/temp_cf/'
lassen_to_uf = rsl_exec + ' ' + in_file + ' ' + temp_file
os.system(lassen_to_uf)
uf_to_cf = radx_exec + ' -const_ngates -params ' + param_file + ' -outdir '+ temp_output + ' -f ' + temp_file
os.system(uf_to_cf)
date_out = os.listdir(temp_output)
file_out = os.listdir(temp_output + date_out[0])
filename = temp_output + date_out[0] + '/' + file_out[0]
print filename
radar = pyart.io.read(filename)
az = radar.azimuth['data'] 
ele = radar.elevation['data'] 
radar.elevation['data'] = az
radar.azimuth['data'] = ele
masked_str = radar.sweep_mode['data'][1,-1]
for line_number in range(len(radar.sweep_mode['data'])):
    for str_loc in range(len(radar.sweep_mode['data'][line_number])):
        radar.sweep_mode['data'][line_number, str_loc] = masked_str
    radar.sweep_mode['data'][line_number, 0] = 'r'
    radar.sweep_mode['data'][line_number, 1] = 'h'
    radar.sweep_mode['data'][line_number, 2] = 'i'


```

```python


# read the data and create the display object
display = pyart.graph.RadarDisplay(radar)

# fields to plot and ranges
fields_to_plot = ['reflectivity_horizontal', 'mean_doppler_velocity']
ranges = [(-32, 64), (-17.0, 17.0)]

# plot the data
nplots = len(fields_to_plot)
plt.figure(figsize=[6 * nplots, 3])

# plot each field
for plot_num in xrange(nplots):
    field = fields_to_plot[plot_num]
    vmin, vmax = ranges[plot_num]

    plt.subplot(1, nplots, plot_num + 1)
    display.plot_rhi(field, 1, vmin=vmin, vmax=vmax, title_flag=False)
    display.set_limits(ylim=[0, 17])

# set the figure title and show
radar_name = display.radar_name
time_text = ' ' + display.time_begin.isoformat() + 'Z '
azimuth = radar.fixed_angle['data'][1]
title = 'RHI ' + radar_name + time_text + 'Azimuth %.2f' % (azimuth)
plt.suptitle(title)
```
