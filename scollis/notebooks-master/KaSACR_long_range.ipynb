```python
import pyart
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.basemap import Basemap
import pickle

%matplotlib inline
```

```python
radar = pyart.io.read('/data/enakasacrppivhD1.a1.20170517.171511.nc')
```

```python
fh = open('/data/ena_terrain.pickle', 'rb')
data_dict = pickle.load(fh)
fh.close()
llcrnrlat=39
urcrnrlon=-27.9
llcrnrlon=-28.06
urcrnrlat=39.1
lats = np.arange(llcrnrlat, urcrnrlat, .02)
lons = np.arange(llcrnrlon, urcrnrlon, .04)
fig = plt.figure(figsize=(12, 6))
m = Basemap(projection='lcc', lon_0=-28, lat_0=39, resolution='c', 
           llcrnrlon=llcrnrlon, llcrnrlat=llcrnrlat,
            urcrnrlon=urcrnrlon, urcrnrlat=urcrnrlat)
xx, yy = m(data_dict['lon'], data_dict['lat'])
```

```python
fig = plt.figure(figsize=(12, 6))
my_display = pyart.graph.RadarMapDisplay(radar)
my_display.plot_ppi_map('reflectivity', basemap = m,
                       lat_lines = lats, lon_lines = lons,
                       vmin = -30, vmax = 60)

CS = m.contour(xx, yy, 
                   data_dict['numpy_dem'].data,
               levels = np.arange(0,400, 100),
               linewidths = 2, cmap=plt.cm.inferno)

plt.clabel(CS, inline=1, fontsize=10, fmt='%1.0f')
plt.savefig('/data/kasacr_vclose.png')
```

```python
llcrnrlat=38.7
urcrnrlon=-27.5
llcrnrlon=-28.5
urcrnrlat=39.5
lats = np.arange(llcrnrlat, urcrnrlat, .1)
lons = np.arange(llcrnrlon, urcrnrlon, .2)
fig = plt.figure(figsize=(12, 6))
m = Basemap(projection='lcc', lon_0=-28, lat_0=39, resolution='c', 
           llcrnrlon=llcrnrlon, llcrnrlat=llcrnrlat,
            urcrnrlon=urcrnrlon, urcrnrlat=urcrnrlat)
xx, yy = m(data_dict['lon'], data_dict['lat'])
fig = plt.figure(figsize=(12, 6))
my_display = pyart.graph.RadarMapDisplay(radar)
my_display.plot_ppi_map('reflectivity', basemap = m,
                       lat_lines = lats, lon_lines = lons,
                       vmin = -30, vmax = 60)

CS = m.contour(xx, yy, 
                   data_dict['numpy_dem'].data,
               levels = np.arange(0,400, 100),
               linewidths = 2, cmap=plt.cm.inferno)

plt.clabel(CS, inline=1, fontsize=10, fmt='%1.0f')
plt.savefig('/data/kasacr_wide.png')
```
