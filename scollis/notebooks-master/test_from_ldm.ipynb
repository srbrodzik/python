```python
import pyart
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
radar = pyart.io.read('/data/testldm.nexrad')
```

```python
display = pyart.graph.RadarDisplay(radar)
fig = plt.figure(figsize = [15,10])
ax = fig.add_subplot(111)
display.plot('reflectivity', 0, vmin=-32, vmax=64.)
display.plot_range_rings([50, 100, 150, 200])
display.plot_cross_hair(5.)
```

```python
lat = 41.7092
lon = -87.9820
max_lat = 43.
min_lat =40
min_lon = -90.5
max_lon = -86
```

```python
display = pyart.graph.RadarMapDisplay(radar)
f = plt.figure(figsize = [18,5])
plt.subplot(1, 3, 1) 
display.plot_ppi_map('differential_reflectivity', max_lat = max_lat, min_lat =min_lat, min_lon = min_lon, max_lon = max_lon,
                     vmin = -1, vmax = 4, lat_lines = np.arange(min_lat,max_lat,.5), lon_lines = np.arange(min_lon, max_lon, 1),
                     resolution = 'l')
plt.subplot(1, 3, 2) 
display.plot_ppi_map('reflectivity',   max_lat = max_lat, min_lat =min_lat, min_lon = min_lon, max_lon = max_lon,
                     vmin = -8, vmax = 64,  lat_lines = np.arange(min_lat,max_lat,.5), lon_lines = np.arange(min_lon, max_lon, 1),
                     resolution = 'l')
plt.subplot(1, 3, 3) 
display.plot_ppi_map('cross_correlation_ratio', max_lat = max_lat, min_lat =min_lat, min_lon = min_lon, max_lon = max_lon,
                     vmin = .7, vmax = 1.,  lat_lines = np.arange(min_lat,max_lat,.5), lon_lines = np.arange(min_lon, max_lon, 1),
                     resolution = 'l')
```
