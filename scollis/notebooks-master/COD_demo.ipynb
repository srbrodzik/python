```python
filename = '/data/KLOT_20150303_1120'
#filename = '/data/KLOT20130418_000354_V06'
radar = pyart.io.read(filename)
#43.6667° N, 70.2667° W
```

```python
myd = pyart.graph.RadarMapDisplay(radar)
fig = plt.figure(figsize = [14,10])
myd.plot_ppi_map( 'reflectivity', vmin = -8, vmax = 54, 
                 resolution = 'l', max_lat = 42.5, 
                 min_lat = 41, min_lon = -89.5, max_lon = -87,
                 embelish = False, shapefile = '/data/ne_50m_lakes')
print("drawn")
m = myd.basemap
m.drawparallels(np.linspace(41, 43, 9),labels=[1,0,0,0])
m.drawmeridians(np.linspace(-88.4, -87, 8),labels=[0,0,0,1])
m.drawrivers()
m.drawcounties()
#m.drawstates()
m.drawmapscale(-88., 41.1, -88., 41.1, 50, barstyle='fancy', fontcolor='k', fillcolor1='b', fillcolor2='k')
myd.plot_point( -87.9706,41.6815, label_text = 'Argonne Lab', label_offset = (0.0,0.0) )
plt.show()
```

```python
rain_z = radar2.fields['reflectivity']['data'].copy()
z_lin = 10.0**(radar2.fields['reflectivity']['data']/10.)
rain_z = (z_lin/300.0)**(1./1.4)  #Z=300 R1.4
radar2.add_field_like('reflectivity', 'rain_z',  rain_z, replace_existing = True)
radar2.fields['rain_z']['units'] = 'mm/h'
radar2.fields['rain_z']['standard_name'] = 'rainfall_rate'
radar2.fields['rain_z']['long_name'] = 'rainfall_rate_from_z'
radar2.fields['rain_z']['valid_min'] = 0
radar2.fields['rain_z']['valid_max'] = 500





```

```python
myd = pyart.graph.RadarMapDisplay(radar2)
fig = plt.figure(figsize = [14,10])
myd.plot_ppi_map( 'rain_z', vmin = 0, vmax = 100, 
                 resolution = 'l', max_lat = 41.8, 
                 min_lat = 41.25, min_lon = -88.3, max_lon = -87.5,
                 embelish = False, shapefile = '/data/ne_50m_lakes')
m = myd.basemap
m.drawparallels(np.linspace(41, 42, 9),labels=[1,0,0,0])
m.drawmeridians(np.linspace(-88.4, -87, 8),labels=[0,0,0,1])
m.drawrivers()
m.drawcounties()
m.drawstates()
m.drawmapscale(-88., 41.55, -88., 41.55, 10, barstyle='fancy', fontcolor='k', fillcolor1='b', fillcolor2='k')
myd.plot_point( -87.9706,41.6815, label_text = 'Argonne Lab', label_offset = (0.0,0.0) )

```
