```python
import pyart
from matplotlib import pyplot as plt
from __future__ import print_function
import numpy as np
%matplotlib inline
```

```python
filename = '/data/xsapr_test_case/XNW130928085001.RAW1HE6'
radar = pyart.io.read(filename)
```

```python
print(radar.fields.keys())
```

```python
lats = np.arange(36,38,.2)
lons = np.arange(-98, -96, .2)
fig = plt.figure(figsize = [16,10])
display = pyart.graph.RadarMapDisplay(radar)
display.plot_ppi_map('reflectivity', sweep = 2, resolution = 'i',
                    vmin = -18, vmax = 64, mask_outside = False,
                    cmap = pyart.graph.cm.NWSRef, 
                    lat_lines = lats, lon_lines = lons)
```
