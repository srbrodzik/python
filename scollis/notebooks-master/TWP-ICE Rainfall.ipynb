```python
R = 300.0 * (myradar.fields['specific_attenuation']['data']) ** 0.89
rainrate = deepcopy(myradar.fields['recalculated_diff_phase'])
rainrate['data'] = np.ma.masked_where(np.logical_or((
    myradar.fields['norm_coherent_power']['data'] < 0.4),
    (myradar.fields['copol_coeff']['data'] < 0.8)), R)
rainrate['valid_min'] = 0.0
rainrate['valid_max'] = 400.0
rainrate['standard_name'] = 'rainfall_rate'
rainrate['long_name'] = 'rainfall_rate'
rainrate['least_significant_digit'] = 1
rainrate['units'] = 'mm/hr'
myradar.fields.update({'rain_rate_A': rainrate})
```

```python
fields_to_plot = ['rain_rate_A', 'recalculated_diff_phase']
ranges = [(0, 180), (-1, 6)]
display = pyart.graph.RadarDisplay(myradar)

nplots = len(fields_to_plot)
plt.figure(figsize=[7 * nplots, 5])
for plot_num in xrange(nplots):
    field = fields_to_plot[plot_num]
    vmin, vmax = ranges[plot_num]
    plt.subplot(1, nplots, plot_num + 1)
    display.plot_ppi(field, 0, vmin=vmin, vmax=vmax, title_flag=False)
    display.plot_range_rings([50,100])
    display.plot_cross_hair(150)
```

```python
def my_qrf(xg,yg,zg):
    hz_dis=np.sqrt(xg**2+yg**2)
    hyp=np.sqrt(hz_dis**2+zg**2)+1.0
    theta=np.arccos(hz_dis/hyp)*180.0/np.pi
    qrf = (zg / 20.0) + np.sqrt(yg ** 2 + xg ** 2) * \
            np.tan(1.0 * 1.5 * np.pi / 180.0) + 100.0  + \
            zg*1000.0*theta**2/(45.0**2)/17000.
    return qrf
```

```python
crud_data=np.where(myradar.fields['corrected_reflectivity_horizontal']['data'] < -500.0)
for field in myradar.fields.keys():
    myradar.fields[field]['data'][crud_data]=-9999
    myradar.fields[field]['data'][crud_data]=np.ma.masked
#cp=corner_to_point([myradar.latitude['data'], myradar.longitude['data']], [cf_lat, cf_lon])
mygrids = pyart.map.grid_from_radars(
    (myradar,),
    grid_shape=(201, 201, 35),
    grid_limits= ((-100.*1000.0, 100.0*1000.0), (-100.0*1000.0, 100.0*1000.0),(0, 17000)),
    fields=myradar.fields.keys(),
    refl_field='corrected_reflectivity_horizontal',
    max_refl=100., qrf_func=my_qrf,copy_field_data=True)
```

```python
eclass = echo_class.steiner(mygrids.axes['x_disp']['data'], 
                 mygrids.axes['y_disp']['data'], 
                 mygrids.axes['z_disp']['data'], 
                 mygrids.fields['corrected_reflectivity_horizontal']['data'],
                 dx=500.0, dy=500.0, intense=45.0, bkg_rad=11000.0, 
                 area_relation='sgp', work_lev=1500.0, peak_relation='sgp', 
                 use_intense=True, fill_val=-9999.0)
```

```python
display = pyart.graph.GridMapDisplay(mygrids, debug=True)

fig = plt.figure(figsize=[15, 8])

# panel sizes
map_panel_axes = [0.05, 0.05, .4, .80]
x_cut_panel_axes = [0.55, 0.10, .4, .30]
y_cut_panel_axes = [0.55, 0.50, .4, .30]
colorbar_panel_axes = [0.05, 0.90, .4, .03]

# parameters
level = 1
vmin = 0.10
vmax = 180
lat = -12.0
lon = 131.0

# panel 1, basemap, radar reflectivity and NARR overlay
ax1 = fig.add_axes(map_panel_axes)
display.plot_basemap(lat_lines=np.arange(-11.5, -13.5, -.2) , 
                     lon_lines=np.arange(129,132,.2), area_thresh=100,
                     resolution='h')
display.plot_grid('rain_rate_A', level=level, vmin=vmin, vmax=vmax)
display.plot_crosshairs(lon=lon, lat=lat)
xd, yd = display.basemap(display.grid_lons, display.grid_lats)
display.basemap.contour(xd, yd, eclass, [0.5], colors=['r'], linewidths=1.)

# colorbar
cbax = fig.add_axes(colorbar_panel_axes)
display.plot_colorbar()

# panel 2, longitude slice.
ax2 = fig.add_axes(x_cut_panel_axes)
display.plot_longitude_slice('rain_rate_A', lon=lon, lat=lat, vmin=vmin, vmax=vmax)
ax2.set_xlabel('Distance from SGP CF (km)')

# panel 3, latitude slice
ax3 = fig.add_axes(y_cut_panel_axes)
display.plot_latitude_slice('rain_rate_A', lon=lon, lat=lat, vmin=vmin, vmax=vmax)

# add a title
slc_height = mygrids.axes['z_disp']['data'][level]
dts = num2date(mygrids.axes['time']['data'], mygrids.axes['time']['units'])
datestr = dts[0].strftime('%H:%M Z on %Y-%m-%d')
title = 'Sliced at ' + str(slc_height) + ' meters at ' + datestr
fig.text(0.5, 0.9, title)
```
