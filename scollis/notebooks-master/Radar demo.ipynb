```python
import pyart
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
myradar=pyart.io.read('/data/KLOT_20150410_0056')
```

```python
my_display = pyart.graph.RadarDisplay(myradar)
tilt = 0
f = plt.figure(figsize=[20,6])
plt.subplot(1,2,1)
my_display.plot_ppi('reflectivity', tilt, vmin=-6, vmax=64.0)
ax = plt.gca()
ax.set_xlim([-60,0])
ax.set_ylim([30,100])
plt.subplot(1,2,2)
my_display.plot_ppi('velocity', 1, vmin=-30, vmax=30)
ax = plt.gca()
ax.set_xlim([-60,0])
ax.set_ylim([30,100])
```
