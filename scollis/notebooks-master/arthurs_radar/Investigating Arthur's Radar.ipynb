```python
import pyart
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
%matplotlib inline
```

```python
filename = '/data/radar/IAC140311172529.RAWZEEJ'
```

```python
radar = pyart.io.read(filename)
```

```python
print radar.fields.keys()
print radar.time['units']
print radar.longitude['data'],radar.latitude['data']
print radar.metadata
for i in range(len(radar.fixed_angle['data'])):
    print "Tilt", i, ": ", radar.fixed_angle['data'][i], radar.fixed_angle['units']
```

```python
font = {'size' : 16}
rc('font', **font)
f = plt.figure(figsize=[15,8])
myd = pyart.graph.RadarMapDisplay(radar)
myd.plot_ppi_map('reflectivity', 0, vmin=-8, vmax=64, resolution = 'l',
                 lat_lines = np.arange(-8,-2,.2),lon_lines = np.arange(-40,-35,.4) )
myd.plot_range_ring(100.0)
myd.plot_range_ring(50.0, line_style = 'k--')
myd.plot_line_xy(np.array([-150000.0, 150000.0]), np.array([0.0,0.0]), line_style = 'k-')
myd.plot_line_xy(np.array([0.0,0.0]),np.array([-150000.0, 150000.0]), line_style = 'k-')


myd.plot_point(radar.longitude['data'], radar.latitude['data'])
ax1 = plt.gca()
x0,x1 = ax1.get_xlim()
y0,y1 = ax1.get_ylim()
ax1.set_aspect((x1-x0)/(y1-y0) - .05)

#plt.text(xp, yp, 'foo')
```
